# Cifar-10 Classification Network

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_training = 49000
num_validation = 50000 - num_training
num_test = 10000

def unpickle(file):
    import sys
    if sys.version_info.major == 2:
        import cPickle
        with open(file, 'rb') as fo:
            dict = cPickle.load(fo)
        return dict['data'], dict['labels']
    else:
        import pickle
        with open(file, 'rb') as fo:
            dict = pickle.load(fo, encoding='bytes')
        return dict[b'data'], dict[b'labels']

def load_train_data():
    data,label = unpickle("cifar-10-batches-py/data_batch_1")
    data = data.reshape((-1,32,32,3),order = 'F')
    for i in range(2,6):
        dataset,labelset = unpickle("cifar-10-batches-py/data_batch_"+str(i))
        dataset = dataset.reshape((-1,32,32,3),order = 'F')
        data = np.concatenate((data,dataset))
        label = np.append(label,labelset)

    data_training= data[:num_training,:,:,:]
    label_training = label[:num_training]
    data_validation = data[num_training:,:,:,:]
    label_validation = label[num_training:]

    return data_training, label_training, data_validation, label_validation

def load_test_data():
    data,label = unpickle("cifar-10-batches-py/test_batch")
    data = data.reshape((-1,32,32,3),order = 'F')
    return data,np.array(label)

# Load cifar-10 data
X_train, Y_train, X_val, Y_val = load_train_data()
X_test, Y_test = load_test_data()

print("Data loaded as follow:\nTraining Input: {} Training Label: {}".format(X_train.shape, Y_train.shape))
print("Validation Input: {} Validation Label: {}".format(X_val.shape, Y_val.shape))
print("Test Input: {} Test Label: {}".format(X_test.shape, Y_test.shape))
```

# Data Augmentation

```python
X_train_aug = X_train[:,::-1,::1]
X_train_ = np.concatenate((X_train,X_train_aug))
Y_train_ = np.append(Y_train,Y_train)

X_train_ = X_train_.astype('float32')/255
X_val_ =X_val.astype('float32')/255
X_test_ = X_test.astype('float32')/255

num_training = num_training*2

print("training dimensions: X: {} Y: {}".format(X_train_.shape,Y_train_.shape))
```

```python
def conv_params(input,stride, kernel, num_filter,num_id):
    stride_shape = [1, stride, stride , 1]
    filter_shape = [kernel,kernel, input.get_shape()[3], num_filter]
    W = tf.get_variable('w'+str(num_id), filter_shape, tf.float32, tf.random_normal_initializer(0.0, 0.02))
    b = tf.get_variable('b'+str(num_id), [1, 1, 1, num_filter], initializer=tf.constant_initializer(0.0))
    return W, b, stride_shape

def fully_connected_params(input_dim, output_dim,num_id):
    filter_shape = [input_dim, output_dim]
    W = tf.get_variable('w'+str(num_id), filter_shape, tf.float32, tf.random_normal_initializer(0.0, 0.02))
    b = tf.Variable(tf.zeros([output_dim]))
    return W,b
```

## Define layers and model

```python
class Network:
    def __init__(self):
        self.num_epoch = 10
        self.batch_size = 128
        self.log_step = 50
        self.beta = 0
        self._build_model()

    def _model(self):
        #Initiate variables for fc5
        self.W_fc5, self.b_fc5 = fully_connected_params(648,256,5)
        
        #Initiate variables for fc6
        self.W_fc6, self.b_fc6 = fully_connected_params(256,10,6)

        #########################################
        #Define each layer afterwards
        #########################################
        
        #Initiate variables for convolution layer 1
        self.W_conv1, self.b_conv1, stride_shape1 = conv_params(self.X,1,7,32,1)
        
        with tf.variable_scope('conv1'):
            self.conv1 = tf.nn.conv2d(self.X, self.W_conv1, stride_shape1, padding='SAME') + self.b_conv1
            self.relu1 = tf.nn.relu(self.conv1)
            self.pool1 = tf.layers.max_pooling2d(self.relu1,3,2)            
            print('conv1 layer: ' + str(self.pool1.get_shape()))
        
        self.batch1 = tf.layers.batch_normalization(self.pool1)
        print('batch_normalization layer: '+str(self.batch1.get_shape()))
        
        #Initiate variables for convolution layer 2
        self.W_conv2, self.b_conv2, stride_shape2 = conv_params(self.batch1,1,5,64,2)

        with tf.variable_scope('conv2'):
            self.conv2 = tf.nn.conv2d(self.batch1, self.W_conv2, stride_shape2, padding='SAME') + self.b_conv2
            self.relu2 = tf.nn.relu(self.conv2)
            self.pool2 = tf.layers.max_pooling2d(self.relu2,3,2)       
            print('conv2 layer: ' + str(self.pool2.get_shape()))
            
        self.dropout = tf.nn.dropout(self.pool2,0.4)
        print ('dropout layer: '+str(self.dropout.get_shape()))
        
        #Initiate variables for convolution layer 3
        self.W_conv3, self.b_conv3, stride_shape3 = conv_params(self.dropout,1,5,64,3)
            
        with tf.variable_scope('conv3'):
            self.conv3 = tf.nn.conv2d(self.dropout, self.W_conv3, stride_shape3, padding='SAME') + self.b_conv3
            print('con3 layer: ' + str(self.conv3.get_shape()))
            
        #Initiate variables for convolution layer 4  
        self.W_conv4, self.b_conv4, stride_shape4 = conv_params(self.conv3,3,3,72,4)
            
        with tf.variable_scope('conv4'):
            self.conv4 = tf.nn.conv2d(self.conv3, self.W_conv4, stride_shape4, padding='SAME') + self.b_conv4 
            self.relu4 = tf.nn.relu(self.conv4)
            print('conv4 layer: ' + str(self.relu4.get_shape()))
            
        self.batch2 = tf.layers.batch_normalization(self.relu4)
        print('batch_normalization layer: '+str(self.batch2.get_shape()))

        self.flat = tf.contrib.layers.flatten(self.batch2)      
        print('flat layer: ' + str(self.flat.get_shape()))

        with tf.variable_scope('fc5'):
            self.fc5 = tf.matmul(self.flat, self.W_fc5) + self.b_fc5
            self.relu5 = tf.nn.relu(self.fc5)
            print('fc5 layer: ' + str(self.relu5.get_shape()))

        with tf.variable_scope('fc6'):
            self.fc6 = tf.matmul(self.relu5, self.W_fc6) + self.b_fc6       
            print('fc6 layer: ' + str(self.fc6.get_shape()))
        
        return self.fc6

    def _input_ops(self):
        # Placeholders
        self.X = tf.placeholder(tf.float32, [None, 32, 32, 3])
        self.Y = tf.placeholder(tf.int64, [None])

    def _build_optimizer(self):
        # Adam optimizer 'self.train_op' that minimizes 'self.loss_op'
        global_step = tf.Variable(0, trainable=False)
        self.learning_rate = tf.train.exponential_decay(1e-3,global_step,500, 0.87,staircase=True)
        self.train_op = tf.train.AdamOptimizer(learning_rate = self.learning_rate).minimize(self.loss_op)
        
    def _loss(self, labels, logits):
        #L2 Regularization
        self.regularizers = tf.nn.l2_loss(self.W_conv1) + tf.nn.l2_loss(self.W_conv2) + \
                            tf.nn.l2_loss(self.W_conv3) + tf.nn.l2_loss(self.W_conv4) + \
                            tf.nn.l2_loss(self.W_fc5) + tf.nn.l2_loss(self.W_fc6)
        
        # Softmax cross entropy loss
        self.loss_op = tf.reduce_mean(
            tf.nn.softmax_cross_entropy_with_logits(labels = labels, logits = logits) + self.beta*self.regularizers)

    def _build_model(self):
        self._input_ops()

        labels = tf.one_hot(self.Y, 10)

        logits = self._model()

        self._loss(labels, logits)
        
        self._build_optimizer()

        predict = tf.argmax(logits, 1)
        correct = tf.equal(predict, self.Y)
        self.accuracy_op = tf.reduce_mean(tf.cast(correct, tf.float32))
        
    def train(self, sess, X_train, Y_train, X_val, Y_val, beta = 0):
        sess.run(tf.global_variables_initializer())
        self.beta = beta

        step = 0
        losses = []
        accuracies = []
        print('-' * 5 + '  Start training  ' + '-' * 5)
        for epoch in range(self.num_epoch):
            print('train for epoch %d' % epoch)
            for i in range(num_training // self.batch_size):
                X_ = X_train[i * self.batch_size:(i + 1) * self.batch_size][:]
                Y_ = Y_train[i * self.batch_size:(i + 1) * self.batch_size]
                feed_dict = {self.X: X_,self.Y: Y_}

                fetches = [self.train_op, self.loss_op, self.accuracy_op]

                _, loss, accuracy = sess.run(fetches, feed_dict=feed_dict)
                losses.append(loss)
                accuracies.append(accuracy)

                if step % self.log_step == 0:
                    print('iteration (%d): loss = %.3f, accuracy = %.3f' %
                        (step, loss, accuracy))
                step += 1
            
            # Print validation results
            print('validation for epoch %d' % epoch)
            val_accuracy = self.evaluate(sess, X_val, Y_val)
            print('-  epoch %d: validation accuracy = %.3f' % (epoch, val_accuracy))
        nploss = np.array(losses)
        npaccuracy = np.array(accuracies)
        plt.figure(1)
        plt.subplot(211)
        plt.plot(nploss)
        
        plt.subplot(212)
        plt.plot(npaccuracy)
        plt.show()

    def evaluate(self, sess, X_eval, Y_eval):
        eval_accuracy = 0.0
        eval_iter = 0
        for i in range(X_eval.shape[0] // self.batch_size):
            X_ = X_eval[i * self.batch_size:(i + 1) * self.batch_size][:]
            Y_ = Y_eval[i * self.batch_size:(i + 1) * self.batch_size]
                        
            feed_dict = {self.X:X_,self.Y:Y_}
            accuracy = sess.run(self.accuracy_op, feed_dict=feed_dict)
            eval_accuracy += accuracy
            eval_iter += 1
        return eval_accuracy / eval_iter
```

```python
# Clear old computation graphs
tf.reset_default_graph()

sess = tf.Session()

model = Network()
model.train(sess, X_train_, Y_train_, X_val_, Y_val,1e-7)
accuracy = model.evaluate(sess, X_test_, Y_test)
print('***** test accuracy: %.3f' % accuracy)
```
